```python
from signalutils.signalutils import readFiles, encodedLabels, calcMfccs, export_to_pickle, unpickle
import numpy as np
```

```python
spk1 = unpickle("data/AM1.pickle")
table_ys= list(map(lambda a: a['y'], spk1))
table_ys.sort()
```

```python
#save data in dicts:    "AM1":{"AM1":0, "AM2:"3",...},"PP":{"AM1":20, ...}...
import os
results = dict()
for path, subdirs, files in os.walk("data"):
        for name in files:
            buffer = unpickle("data/"+name)
            for idx, element in enumerate(buffer):
                if idx == 0:
                    results[element['x']] = dict()
                results[element['x']][element['y']] = element['cost']
```

```python
# clear "AM1", "AP_1", "as2", "PP-2" to XX_N
def normalizeName(name):
    import re
    name = re.sub(r'[^\w]', '', name)
    name = name.replace("_", "")
    recordName = name[:2]
    recordNumber= name[2:]
    return f"{recordName.upper()}_{int(recordNumber)}"
```

```python
#transform data to {"XX": [cost1, cost2,...], "XY", [cost1, cost2, ...]...}
tableData = {}
for x in table_ys:
    buff = []
    for y in table_ys:
        buff.append(results[x][y])
        tableData[normalizeName(x)] = buff
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = tableData, index = list(map(lambda labelName: normalizeName(labelName), table_ys)))
```

```python
#normalize costs
column_maxes = df.max()
df_max = column_maxes.max()
df = df / df_max
```

```python
speakers = set(map(lambda el: el.split("_")[0],df.columns))
speakers
```

```python
['AB' for x in range(0,3)]
```

```python
allCorrectlyPredicted = 0
trueValues = []
predictedValues = []
for testedSpeaker in speakers:
    n_smallest_distances = 8
    topMinDistances = df.nsmallest(n_smallest_distances, [f"{testedSpeaker}_1"])[f"{testedSpeaker}_1"].index
    correctlyPredictedSamples = sum(map(lambda element: element.split("_")[0]==testedSpeaker, topMinDistances))
    
    trueValues = trueValues + [testedSpeaker for x in range(0,n_smallest_distances)]
    predictedValues = predictedValues + list(map(lambda element: element.split("_")[0], topMinDistances))

    print(f"[{testedSpeaker}]: {correctlyPredictedSamples}/{n_smallest_distances} = {(correctlyPredictedSamples/n_smallest_distances)*100}% (ACC)")
    allCorrectlyPredicted = allCorrectlyPredicted + correctlyPredictedSamples

acc = allCorrectlyPredicted / (8*len(speakers))
print(f"Overall accuracy: {acc*100}%")
```

```python
signals, labelEncoder = encodedLabels(readFiles(path="../samples"))
```

```python
from sklearn.metrics import confusion_matrix

#Get the confusion matrix
cf_matrix = confusion_matrix(labelEncoder.transform(trueValues), labelEncoder.transform(predictedValues))

print(cf_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(20,16))


ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Macierz pomyÅ‚ek\n\n');
ax.set_xlabel('\n Przewidziany mÃ³wca')
ax.set_ylabel('WÅ‚aÅ›ciwy mÃ³wca ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(labelEncoder.classes_)
ax.yaxis.set_ticklabels(labelEncoder.classes_)
## Display the visualization of the Confusion Matrix.
plt.show()

confusionfigure = ax.get_figure()
confusionfigure.savefig('conf.png')
```
